# CreateControlDatasets
### Creates the Similar and Dissimilar triplets where the Dissimilar triplets have random sequences as their negatives

## Imports

```python
from FantomControl import ControlDriver
import os
```

## Files

```python
ctrl_list = [("LGNR", "length_gc_no_repeats"), ("LGR", "length_gc_repeats"), ("LNR", "length_no_repeats"), ("LR", "length_repeats")]

fasta_dir = "../Data/HG38/Chromosomes/"
data_dir = "../Data/Datasets/"
fantom_dir = f'../Data/FANTOM/'

set_list = ['train', 'valid', 'test']
```

## Parameters

```python
cores = 8
iterations = 500
```

## Running

```python
for ctrl in ctrl_list:    
    for x in set_list:
        y = f'{fantom_dir}/{x}'
        z = f'{data_dir}/{ctrl[0]}/{x}'

        fantom_file = f'{y}.usage.matrix'
        pickle_file = f'{y}_crm.pickle'
        sim_file = f'{z}_triplet_sim.npy'
        dis_file = f'{z}_triplet_dis.npy'
        seq_file = f'{z}_sequences.fa'
        ctrl_file = f'{data_dir}/{ctrl[0]}/{x}_{ctrl[1]}.fa'
        
        ControlDriver(fantom_file, pickle_file, sim_file, dis_file, fasta_dir, seq_file, ctrl_file, cores, iterations).run()
        
```
