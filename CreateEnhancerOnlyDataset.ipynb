# CreateEnhancerOnlyDataset
### Creates the similar triplets and dissimilar triplets for enhancers

## Imports

```python
from Fantom import Driver
import os
```

## Files

```python
fasta_dir = "../Data/HG38/Chromosomes/"
data_dir = "../Data/Datasets/"
fantom_dir = f'../Data/FANTOM/'
out_dir = f'{data_dir}/Enhancer'
```

```python
fantom_list = []
pickle_list = []
sim_list = []
dis_list = []
seq_list = []

for x in ['train', 'valid', 'test']:
    y = f'{fantom_dir}/{x}'
    z = f'{out_dir}/{x}'
    fantom_list.append(f'{y}.usage.matrix')
    pickle_list.append(f'{y}_crm.pickle')
    sim_list.append(f'{z}_triplet_sim.npy')
    dis_list.append(f'{z}_triplet_dis.npy')
    seq_list.append(f'{z}_sequences.fa')
```

## Parameters

```python
cores = 8
iterations = 500
```

## Running

```python
for i in range(len(fantom_list)):
    Driver(fantom_list[i], pickle_list[i], sim_list[i], dis_list[i], fasta_dir, seq_list[i], cores, iterations).run()
```
